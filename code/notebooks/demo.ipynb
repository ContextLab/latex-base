```python
import numpy as np
import seaborn as sns

import os
from pathlib import Path
from matplotlib import pyplot as plt
```

# Create Figure 1

Save under ../../papers/figs/source/sin.pdf and ../../papers/figs/source/cos.pdf

```python
figdir = Path.joinpath(Path(os.path.dirname(os.path.realpath("__file__"))).parent.parent,
                       'paper', 'figs', 'source')
sin_fname = str(Path.joinpath(figdir, 'sin.pdf'))
cos_fname = str(Path.joinpath(figdir, 'cos.pdf'))
```

## Panel A

Create a sine wave over the interval $x \in [0, 4\pi]$.

```python
x = np.linspace(0, 4 * np.pi, 1000)
plt.plot(x, np.sin(x), 'k', linewidth=2);
plt.xlabel('$x$', fontsize=14);
plt.ylabel('$\sin(x)$', fontsize=14);
plt.savefig(sin_fname, transparent=True, bbox_inches='tight')
```

## Panel B

Create a cosine wave over the interval $x \in [0, 4\pi]$.

```python
plt.plot(x, np.cos(x), 'k', linewidth=2);
plt.xlabel('$x$', fontsize=14);
plt.ylabel('$\cos(x)$', fontsize=14);
plt.savefig(cos_fname, transparent=True, bbox_inches='tight')
```

# Note

The final figure, ../../papers/figs/trig.pdf uses linked source files.  Opening trig.pdf in Adobe Illustrator (or another program that can read Illustrator-compatible PDFs) will draw from the relevant source files (i.e., source/sin.pdf and source/cos.pdf) such that updating the source files will change the final figure.
